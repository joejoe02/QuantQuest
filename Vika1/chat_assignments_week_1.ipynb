```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

data = yf.download(["AAPL", "MSFT"], start='2023-01-01', end='2024-01-01')
```

```python
data.columns = ['_'.join(col) if isinstance(col, tuple) else col for col in data.columns]
```

```python
plt.figure(figsize=(10, 6))

plt.plot(data.index, data['Close_AAPL'], label='AAPL Close', marker='o')
plt.plot(data.index, data['Close_MSFT'], label='MSFT Close', marker='o')

plt.title('AAPL vs MSFT Closing Prices (Monthly)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
